# **WEB SCRAPING com PYTHON**
 
Como fazer um web scraping do site rotten tomatoes 

```
#importando as bibliotecas
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import csv 
import pandas as pd
```

```
#inserindo a url da pagina de destino, numero de 23 é o numero de paginas que quero coletar
rotten_urls=['https://www.rottentomatoes.com/m/star_wars_the_rise_of_skywalker/reviews?type=&sort=']*23
```

```
rotten_urls
```

```
#criando lista de urls 
#
page=2
while (page<=23):
    url_pos = page-1
    rotten_urls[url_pos] = rotten_urls[url_pos]+'&page='+ str(page)
    page = page + 1
```

```
rotten_urls
```

```
rotten_urls[url_pos]
```

```
# coletando as reviews do site

reviews=[]
for url in rotten_urls:
    uClient = uReq(url)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html,"html.parser")
    reviews = reviews + [review.text for review in page_soup.find_all("div", {"class": "the_review"})]
```

# **WORLD CLOUD em PYTHON:**
 **Como fazer uma nuvem de palavra em Python?** 


```
##CONSTRUINDO A NUVEM DE PALAVRA
#importando as bibliotecas necessarias
import csv 
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from wordcloud import WordCloud, ImageColorGenerator
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```
#usar o token nos dados 
#token separa as palavras da frase "Eu amo pizza", "Eu" "amo" "pizza", vale ressaltar que separa as palavras com significado tipo "guarda-chuva"

stop_words=set(stopwords.words('english')) #lista de stopword em ingles, stopw word palavras que não traz significado para o texto como: para, a, também,o
stop_words#lista de stop word da biblioteca nltk

# tokenização das reviews, "separando" as palavras 
word_tokens=[]
for review in reviews:
  word_tokens=word_tokens+word_tokenize(review)

filtered_word = [w for w in word_tokens if not w in stop_words]#filtrando, eliminando as stop words das revies
```

```
##Entendendo o que foi feito
print(stop_words) 
print(word_tokens[:15])
print(filtered_word[:13])
```

```
# concatenar as palavras
all_summary = " ".join(s for s in filtered_word)

# ver quantidade de palavras
print("Quantidade de Palavras: {}".format(len(all_summary)))


```

```
all_summary
```

```
# gerar uma wordcloud
wordcloud = WordCloud(background_color="black").generate(all_summary)
```

```
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("wordcloud.png")
```

```
##Considerar apenas adjetivos
##JJ como o pacote categoriza adj
adjs=[word for (word,pos) in nltk.pos_tag(filtered_word)if(pos[:2]== 'JJ')]
print(adjs)
```

```
# concatenar as palavras
palavras = " ".join(s for s in adjs)
# gerar uma wordcloud
wordcloud = WordCloud(background_color="black").generate(palavras)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud);
wordcloud.to_file("wordcloud.png")
```
